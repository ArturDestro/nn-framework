```python
def numerical_gradient(model, layer, x, y, loss_fn, eps=1e-5):
    num_grad = np.zeros_like(layer.w)

    for i in range(layer.w.shape[0]):
        for j in range(layer.w.shape[1]):
            original = layer.w[i, j]

            # w + eps
            layer.w[i, j] = original + eps
            y_pred = model.forward(x)
            loss_plus = loss_fn.forward_pass(y_pred, y)

            # w - eps
            layer.w[i, j] = original - eps
            y_pred = model.forward(x)
            loss_minus = loss_fn.forward_pass(y_pred, y)

            num_grad[i, j] = (loss_plus - loss_minus) / (2 * eps)

            layer.w[i, j] = original  # restaura

    return num_grad
```

```python
def relative_error(g1, g2):
    return np.linalg.norm(g1 - g2) / (np.linalg.norm(g1) + np.linalg.norm(g2) + 1e-8)
```

```python
np.random.seed(0)

x = np.random.randn(1, 3)
y = np.random.randn(1, 2)

model = Model()
layer = Dense(3, 2)
model.add(layer)

loss_fn = MSELoss()


# Forward
y_pred = model.forward(x)
loss = loss_fn.forward_pass(y_pred, y)

# Backward
grad_loss = loss_fn.backward_pass()
model.backward(grad_loss)

grad_analitico = layer.dw.copy()

# NUMÉRICO POR ÚLTIMO
grad_numerico = numerical_gradient(
    model, layer, x, y, loss_fn
)
```

```python
def relative_error(g1, g2):
    return np.linalg.norm(g1 - g2) / (np.linalg.norm(g1) + np.linalg.norm(g2) + 1e-8)
```

```python
erro = relative_error(grad_analitico, grad_numerico)
print("Erro relativo:", erro)
```
