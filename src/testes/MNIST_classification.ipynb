```python
from sklearn.datasets import fetch_openml
```

```python
import numpy as np
from framework.model import *
from framework.layers import *
```

```python
X, y = fetch_openml("mnist_784", version=1, return_X_y=True, as_frame=False)
X = X.astype(np.float32) / 255.0
y = y.astype(np.int64)
X = X.reshape(-1, 28, 28, 1)
numclasses=10
one_hot = np.eye(numclasses)[y]
```

```python
X_train, X_test = X[:60000], X[60000:]
y_train, y_test = one_hot[:60000], one_hot[60000:]
```

```python
init = XavierNormal()
```

```python
np.random.seed(0)

model = Model()


model.add(Flatten())
model.add(Dense(784, 128, initializer=init))
model.add(ReLU())
model.add(Dropout(drop_rate=0.5))
model.add(Dense(128, 10, initializer=init))
```

```python
loss=CrossEntropyLoss()
op = SGD(lr=0.05)
```

```python
model.fit(X_train, y_train, loss, op, epochs=100, batch_size=32, validation_split=0.2, validation_data=None, shuffle=True, verbose=True)
```

```python
metrics = [Accuracy()]
```

```python
results = model.evaluate(X_test, y_test, metrics)
```
